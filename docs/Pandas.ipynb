## 凸优化

```python
import scipy.optimize as sco
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

noise = np.random.normal(0, 0.4, 100)
x = np.linspace(0, 10, 100)
y = np.sin(x) + noise
f = interp1d(x, y)
f2 = interp1d(x, y, kind='cubic')
xm = (x[10] + x[11])/2
xnew = np.linspace(0,10,10)
xnew2 = np.linspace(0,10,10)
plt.plot(x,y,'o',xnew, f(xnew),'-',xnew,f2(xnew),'--',xnew,np.sin(xnew),linewidth=1)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(0,10,.5)
y = np.arange(-10,10,.5)
x_grid,y_grid = np.meshgrid(x,y)
z_grid = x_grid**2 + y_grid**2
ax.plot_surface(x_grid,y_grid,z_grid,rstride=1,cstride=1,cmap='hot')
plt.show()
```
